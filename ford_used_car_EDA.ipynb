#### Predict Price of the Used Cars

##### Import required library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### importing data file

```python
ford = pd.read_csv('ford.csv')
# display(ford.head(5))
# display(ford.tail(5))
# ford.info()
display(ford.shape)
print('type of ford = ',type(ford))
```

##### Exploratory Data Analysis (EDA) 
- using groupby/pivot_table and barplot (total 9 barplots), 
- based on model, transmission, and fuelType

```python
ford_model=ford.groupby('model')['price'].agg(['count','mean','sum'])
ford_model['Percentage'] = (ford_model['count']/ford_model['count'].sum())*100
display(ford_model)

ford_transmission=ford.groupby('transmission')['price'].agg(['count','mean','sum'])
ford_transmission['Percentage'] = (ford_transmission['count']/ford_transmission['count'].sum())*100
display(ford_transmission)

ford_fuelType=ford.groupby('fuelType')['price'].agg(['count','mean','sum'])
ford_fuelType['Percentage'] = (ford_fuelType['count']/ford_fuelType['count'].sum())*100
display(ford_fuelType)
```

```python
# display(ford_model.index)
ford_model = ford_model.sort_values(by='count', ascending=False)
ford_transmission = ford_transmission.sort_values(by='count', ascending=False)
ford_fuelType = ford_fuelType.sort_values(by='count', ascending=False)
# display(ford_model['Percentage'])
# display(ford_transmission['Percentage'])
# display(ford_fuelType['Percentage'])
```

```python
plt.figure(figsize=(20,16))

plt.subplot(3, 3, 1)
sns.barplot(x = 'Percentage', y=ford_model.index, data=ford_model, orient='h', color='blue')
plt.title("Percentage of cars in each model")
plt.xlabel("Percentage of cars (%)")
plt.ylabel('ford models')
plt.grid()

plt.subplot(3, 3, 2)
sns.barplot(x='mean', y=ford_model.index, data=ford_model, orient='h', color='orange')
plt.title("Average resale value of the cars of each model")
plt.xlabel("Average price ($)")
plt.ylabel("Ford Models")
plt.grid()

plt.subplot(3, 3, 3)
sns.barplot(x='sum', y=ford_model.index, data=ford_model, orient='h', color='green')
plt.title("total resale value of the cars by each model")
plt.xlabel("Total price in $")
plt.ylabel(" Ford Models")
plt.grid()

plt.subplot(3, 3, 4)
sns.barplot(x = 'Percentage', y=ford_transmission.index, data=ford_transmission, orient='h', color='blue')
plt.title("Percentage of cars in each transmission")
plt.xlabel("Percentage of cars (%)")
plt.ylabel('ford transmission')
plt.grid()

plt.subplot(3, 3, 5)
sns.barplot(x='mean', y=ford_transmission.index, data=ford_transmission, orient='h', color='orange')
plt.title("Average resale value of the cars of each transmission")
plt.xlabel("Average price ($)")
plt.ylabel("Ford transmission")
plt.grid()

plt.subplot(3, 3, 6)
sns.barplot(x='sum', y=ford_transmission.index, data=ford_transmission, orient='h', color='green')
plt.title("total resale value of the cars by each transmission")
plt.xlabel("Total price in $")
plt.ylabel(" Ford transmission")
plt.grid()

plt.subplot(3, 3, 7)
sns.barplot(x = 'Percentage', y=ford_fuelType.index, data=ford_fuelType, orient='h', color='blue')
plt.title("Percentage of cars in each fuelType")
plt.xlabel("Percentage of cars (%)")
plt.ylabel('ford fuelType')
plt.grid()

plt.subplot(3, 3, 8)
sns.barplot(x='mean', y=ford_fuelType.index, data=ford_fuelType, orient='h', color='orange')
plt.title("Average resale value of the cars of each fuelType")
plt.xlabel("Average price ($)")
plt.ylabel("Ford fuelType")
plt.grid()

plt.subplot(3, 3, 9)
sns.barplot(x='sum', y=ford_fuelType.index, data=ford_fuelType, orient='h', color='green')
plt.title("total resale value of the cars by each fuelType")
plt.xlabel("Total price in $")
plt.ylabel(" Ford fuelType")
plt.grid()

plt.suptitle("Barplot of the Ford used car")
plt.tight_layout()
plt.show()
```

#### Pivot table

```python
ford_pivot=pd.pivot_table(ford, values='price', index = ['model','transmission','fuelType'], aggfunc=['count','mean','sum'])
display(ford_pivot)
```
